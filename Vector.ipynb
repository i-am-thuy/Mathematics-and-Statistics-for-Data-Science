```python
#Import các thư viện cần thiết
import numpy as np
import pandas as pd
from numpy.linalg import norm
import matplotlib.pyplot as plt
from sklearn import decomposition
from sklearn.preprocessing import scale
import scipy
from sklearn.decomposition import PCA
```

```python
### Cau 1 - Vector ###
#Tao cac diem co toa do da cho
A = np.array([20,10])
B = np.array([70,30])
C = np.array([50,55])
dist_AB = norm(A-B)
dist_AC = norm(A-C)
print("Khoang cach tu A den B: ", dist_AB)
print("Khoang cach tu A den C: ", dist_AC)
```

Để tính khoảng cách giữa hai điểm trong không gian toạ độ Oxy, ta sẽ tìm độ dài vector nối hai điểm ấy.

Vector trong toạ độ Oxy có độ lớn được tính theo 3 cách, trong đó L2, hay còn được gọi là khoảng cách Euclid, là thước đo phổ biến để tính khoảng cách giữa hai điểm.

Vì khoảng cách Euclid từ A đến B nhỏ hơn khoảng cách từ A đến C. Như vậy, C gần A hơn B

```python
#Tinh quang duong toi uu tu diem A den diem B
print("Quang duong di ngan nhat tu A den B: ", norm(A-B, np.inf))
```

Quãng đường di chuyển tối ưu giữa hai điểm trong trục toạ độ Oxy có thể hiểu là số bước đi ngắn nhất để đi từ điểm này qua điểm kia. Theo đó, độ lớn vector được tính theo norm L3 là cách để đo lường quãng đường này, và 50 là kết quả tính được.
